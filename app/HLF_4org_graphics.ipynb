```python
import matplotlib.pyplot as plt

concurrency = [5, 20, 50, 100]

throughput_data = {
    'one org': [4.95, 19.85, 49.65, 61.72],
    'two orgs': [4.96, 19.83, 47.09, 55.18],
    'three orgs': [4.96, 19.80, 46.55, 47.50],
    'four orgs': [4.96, 19.83, 45.13, 44.15]
}

latency_data = {
    'one org': [0.14, 0.14, 0.14, 0.17],
    'two orgs': [0.15, 0.16, 0.17, 0.26],
    'three orgs': [0.16, 0.16, 0.20, 0.33],
    'four orgs': [0.17, 0.17, 0.25, 0.42]
}

# Configuración de estilos
line_styles = ['-', ':', '--', '-']
markers = [None, None, None, '*']
labels = ['one org', 'two orgs', 'three orgs', 'four orgs']

def crear_grafica_tesis(data, ylabel, filename):
    plt.figure(figsize=(8, 5))

    for i, label in enumerate(labels):
        plt.plot(concurrency, data[label],
                 label=label,
                 color='black',
                 linestyle=line_styles[i],
                 marker=markers[i],
                 linewidth=2,
                 markersize=10 if markers[i] == '*' else 7,
                 markeredgecolor='black')

    plt.xlabel('Carga (TPS)', fontweight='bold')
    plt.ylabel(ylabel, fontweight='bold')
    plt.xticks(concurrency)
    plt.grid(axis='y', linestyle='-', alpha=0.2)


    leg = plt.legend(title='Endorsement', frameon=False, loc='upper left')
    plt.setp(leg.get_title(), fontweight='bold')

    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.show()

# 4. Generar archivos PDF
crear_grafica_tesis(throughput_data, 'Throughput (TPS)', 'HLF_throughput_4ORGS.png')
crear_grafica_tesis(latency_data, 'Latencia (s)', 'HLF_latency_4ORGS.png')
```
